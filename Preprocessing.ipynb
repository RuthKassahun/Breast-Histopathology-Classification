### **Load Data**

```python
clahe = cv2.createCLAHE(clipLimit = 3.0, tileGridSize = (8,8))
```

```python
def process_data(tag='train'):
  num = 0
  tag_dir = os.path.join(image_dir, tag)
  tag_path = Path(tag_dir)
  for mag_dir in tag_path.iterdir():  
    mag_label = mag_dir.stem
    print("*", mag_label)
    for img_path in mag_dir.glob('*.png'):
      img_label = img_path.stem
      #print(img_label)
      img = cv2.imread(img_path.as_posix(), 1)
      norm_img = cv2.normalize(img,  None, 0, 255, cv2.NORM_MINMAX)
      denoised_img = cv2.fastNlMeansDenoisingColored(norm_img, None, 3, 3, 7, 21)
      lab_image = cv2.cvtColor(denoised_img, cv2.COLOR_BGR2LAB)
      l,a,b = cv2.split(lab_image)
      clahe_l = clahe.apply(l)
      clahe_lab_img = cv2.merge((clahe_l, a, b))
      clahe_bgr_img = cv2.cvtColor(clahe_lab_img, cv2.COLOR_LAB2BGR)
      cv2.imwrite(f'/gdrive/My Drive/Project/Denoised_CLAHE_Cl3/{tag}/{mag_label}/{img_label}.png',clahe_bgr_img)
      num = num + 1

  print(num)
```

```python
# train images
process_data('train')
# test images
#process_data('test')
```
